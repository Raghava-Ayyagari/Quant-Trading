```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

A=np.array(
    [[1,0.5],
     [0.3,1]]
)
n=100000
start=[0,0]
X=np.random.randn(2,n)
b=(np.array([0,0]).T).reshape((2,1))
Y=A.T@X+b
print(np.shape(Y))
x=Y[0]
y=Y[1]
indexes=np.arange(n)
Pdf=(1/(2*np.pi*np.linalg.det(A)))*np.exp(np.array([-0.5*(np.array([Y[0][i],Y[1][i]]).T@A@A.T@np.array([Y[0][i],Y[1][i]])) for i in range(n)]))
plt.scatter(x,y,c=Pdf,cmap="viridis",alpha=0.5,s=0.8)
plt.colorbar()
plt.scatter(b[0]*indexes,b[1]*indexes,c="red",s=0.5)
plt.show()
```

```python
Z=np.zeros((2,n))
Z[0]=np.cumsum(Y[0])
Z[1]=np.cumsum(Y[1])
plt.scatter(Z[0],Z[1],c=indexes,cmap="viridis",alpha=0.5,s=0.75)
plt.colorbar()
plt.scatter(b[0]*indexes,b[1]*indexes,c="red",s=0.5)
plt.show()
```
