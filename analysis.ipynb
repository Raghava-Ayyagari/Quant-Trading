```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
```

```python
Data=yf.download("^BSESN",start="2021-01-01",end="2026-01-01")
plot=Data.copy()
plt.plot(np.linspace(0,len(plot),len(plot)),plot['Close'])
plt.show()
```

```python
Data['logreturns']=np.log(Data['Close']/Data['Close'].shift(1))
plt.scatter(np.linspace(0,len(Data),len(Data)),Data['logreturns'],s=0.75)
plt.show()
```

```python
Data['logreturns']=np.log(Data['Close']/Data['Close'].shift(1))
plt.plot(np.linspace(0,len(Data),len(Data)),Data['logreturns'])
plt.show()
```

```python
print(Data['logreturns'].mean(),Data['logreturns'].std())
plt.hist(Data['logreturns'],bins=100,density=True)
plt.show()
```
